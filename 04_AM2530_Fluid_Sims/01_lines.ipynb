# Defining the class System, and System properties for a point
A few system properties can be calculated at a point.\
For eg,
## 1, Translation rate of a fluid element/particle
Translation rate, or $\vec a = \dot{\vec{r}}$ is caused due to two things:
1. Time dependence of velocity, $\dfrac{\partial v}{\partial t}$
2. Gradient of this velocity, $(\vec{v} \cdot \vec{\nabla}) \vec{v}$
These two conclusions are derived from the formula
$$\vec a = \dot{\vec{r}} = \dfrac{\partial v}{\partial t} + \sum{\dfrac{\partial \vec{v}}{\partial x} \times \dfrac{\partial x}{\partial t}}\\
\implies \vec a = \dfrac{\partial v}{\partial t} + \sum{u \times \dfrac{\partial \vec{v}}{\partial x}}\\
= \dfrac{\partial \vec{v}}{\partial t} + (\vec{v} \cdot \vec{\nabla}) \vec{v}
$$

## 2. Roatation rate
This is simply denoted by the formula
$$\omega = 1/2(\vec{\nabla} \times \vec{v})$$

## 3. Deformation rate
Mathematically
$$\epsilon_{ij} = 1/2(\vec{\nabla} \vec{v} + (\vec{\nabla} \vec{v})^T)\\$$
where $\vec{a} \vec{b}$ is the outer product of a and b
$$\\
= \begin{bmatrix}
u_x & u_y + v_x & u_z + w_x \\
u_y + v_x & v_y & v_z + w_y \\
u_z + w_x & v_z + w_y & w_z
\end{bmatrix}$$
where $a_b = \dfrac{\partial a}{\partial b}$
