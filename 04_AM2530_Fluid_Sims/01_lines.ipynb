# Theory
A few system properties can be calculated at a point.\
For eg,
## 1. Translation rate of a fluid element/particle
Translation rate, or $\vec a = \dot{\vec{r}}$ is caused due to two things:
1. Time dependence of velocity, $\dfrac{\partial v}{\partial t}$
2. Gradient of this velocity, $(\vec{v} \cdot \vec{\nabla}) \vec{v}$
These two conclusions are derived from the formula
$$\vec a = \dot{\vec{r}} = \dfrac{\partial v}{\partial t} + \sum{\dfrac{\partial \vec{v}}{\partial x} \times \dfrac{\partial x}{\partial t}}\\
\implies \vec a = \dfrac{\partial v}{\partial t} + \sum{u \times \dfrac{\partial \vec{v}}{\partial x}}\\
= \dfrac{\partial \vec{v}}{\partial t} + (\vec{v} \cdot \vec{\nabla}) \vec{v}
$$

## 2. Roatation rate
This is simply denoted by the formula
$$\omega = 1/2(\vec{\nabla} \times \vec{v})$$

## 3. Deformation rate
Mathematically
$$\epsilon_{ij} = 1/2(\vec{\nabla} \vec{v} + (\vec{\nabla} \vec{v})^T)\\$$
where $\vec{a} \vec{b}$ is the outer product of a and b
$$\\
= \begin{bmatrix}
u_x & u_y + v_x & u_z + w_x \\
u_y + v_x & v_y & v_z + w_y \\
u_z + w_x & v_z + w_y & w_z
\end{bmatrix}$$
where $a_b = \dfrac{\partial a}{\partial b}$

## 1. The `math_helper` class
The `math_helper` class contains function that are required later to solve the tough theoretical maths 

```python
from sympy.vector import curl, divergence, gradient, laplacian
class math_helper:
    def smpMatTosmpVec(self, SMMatrix, SMCoordSys):
        return smp.vector.matrix_to_vector(smp.Matrix(SMMatrix), SMCoordSys).replace(x, SMCoordSys.x).replace(y, SMCoordSys.y)
    def formatvector(self, SMVec, SMCoordSys):
        return [SMVec[0].replace(x, SMCoordSys.x).replace(y, SMCoordSys.y), SMVec[1].replace(x, SMCoordSys.x).replace(y, SMCoordSys.y)]
    def divergence(self, SMVec, SMCoordSys):
        return divergence(SMVec).doit()
    def curl(self, SMVec, SMCoordSys):
        return curl(SMVec).doit()

    def outer_prod(self, Vec1: smp.vector.vector.VectorAdd, Vec2: smp.vector.vector.VectorAdd):
        return Vec1.outer(Vec2)
# Initialise the function
fmath = math_helper()
```
